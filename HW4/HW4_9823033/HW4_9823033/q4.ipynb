```python
#p9
data3=pd.read_csv("C:/Users/acer/Desktop/HW4/data.csv")
data3

cols3=data3.columns[0:5]
cols3
data3=data3[cols3]
data3
```

```python
#p9
for column_name in cols3:
    data3[column_name] = data3[column_name].replace(0, np.nan)

data3.isnull().sum()
```

```python
#p9
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(data3)
data3= imputer.fit_transform(data3)
data3=pd.DataFrame(data3)
data3
data3.isnull().sum()
```

```python
#p10
imputer = SimpleImputer(missing_values=np.nan, strategy='median')
imputer.fit(data3)
data3= imputer.fit_transform(data3)
data3=pd.DataFrame(data3)
data3
#data3.isnull().sum()
```

```python
#p10
imputer = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
imputer.fit(data3)
data3= imputer.fit_transform(data3)
data3=pd.DataFrame(data3)
data3
```

```python
#p10
imputer = SimpleImputer(missing_values=np.nan, strategy='constant',fill_value=58)
imputer.fit(data3)
data3= imputer.fit_transform(data3)
data3=pd.DataFrame(data3)
data3
```
